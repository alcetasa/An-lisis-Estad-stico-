## Distribución normal





Fuente: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multivariate_normal.html#scipy.stats.multivariate_normal

```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.stats
```

```python
from scipy.stats import norm
```

```python
norm.cdf(0)
```

```python
norm.rvs(30,4,size=10)
```

```python
norm.rvs(size=100)
```

```python
norm.cdf(18,20,2)
```

```python
norm.ppf(0.15865525393145707,loc=20,scale=2)
```

```python
norm.ppf(.75,20,5)
```

```python
norm.rvs(20,5,100)
```

# Distribución Gamma




```python
from scipy.stats import gamma
fig, ax = plt.subplots(1,1)
a=50
b=3
x = np.linspace(gamma.ppf(0.0001, a),
                gamma.ppf(0.9999, a), 100)
ax.plot(x, gamma.pdf(x, a))
```

```python
r=gamma.rvs(10,4,size=30)
r
```

```python
gamma.cdf(10,3,5)
```

```python
gamma.pdf(10,3,5)
```

```python
gamma.ppf(0.8753479805169189,3,5)
```

# Distribución Uniforme



Fuente: https://docs.scipy.org/doc/scipy/reference/stats.html

```python
from scipy.stats import uniform
fig, ax = plt.subplots(1,1)
a=10
b=30
x = np.linspace(uniform.ppf(0.0001,a,b),
                uniform.ppf(0.9999, a,b), 100)
ax.plot(x, uniform.pdf(x, a,b))
```

```python
uniform.cdf(25,10,30)
```

```python
uniform.pdf(25,10,30)
```

```python
uniform.ppf(0.5,10,30)
```
